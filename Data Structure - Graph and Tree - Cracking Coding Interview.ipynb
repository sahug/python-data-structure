**Route Between Nodes**: Given a directed graph, design an algorithm to find out whether there is a route between two nodes.

image.png

```python
routes = [
      ("Mumbai", "Paris"),
      ("Mumbai", "Dubai"),
      ("Paris", "Dubai"),
      ("Paris", "New York"),
      ("Dubai", "New York"),
      ("New York", "Toronto"),
  ]
```

```python
class Graph:

  def __init__(self, edges):
    self.edges = edges

    self.routes_dict = {}
    for start, end in self.edges:
      if start in self.routes_dict:
        self.routes_dict[start].append(end)        
      else:
        self.routes_dict[start] = [end]
    
    print(self.routes_dict)
    
  def route(self, start, dest):
    if dest in self.routes_dict[start]:
      return True
    else:
      return False
```

```python
graph = Graph(routes)
```

```python
print(graph.route("Mumbai", "Dubai"))
print(graph.route("New York", "Mumbai"))
```
